```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
import math
```

```python
print(os.getcwd())
```

```python
dados = pd.read_excel('Titanic.xlsx')
```

```python
dados.head()
```

```python
d=dados.loc[:,['Survived','Pclass','Sex','Age','SibSp','Parch','Fare']]
d.columns=['Sobreviveu','Classe','Sexo','Idade','Companheiro_ou_Irmão_a_bordo', 'Pais_ou_filhos_a_bordo','Preço']
d.head()
```

```python
d_sex = pd.get_dummies(d['Sexo'])
ds = pd.concat([d, d_sex], axis=1)

d_class = pd.get_dummies(d['Classe'])
dc = pd.concat([ds, d_class], axis=1)
dc.head()
```

```python
dc.Companheiro_ou_Irmão_a_bordo.describe()
ut1 = dc.Companheiro_ou_Irmão_a_bordo.value_counts(sort=False).reindex()
print(ut1,'\n')
```

```python
dc.loc[dc.Companheiro_ou_Irmão_a_bordo==0, 'CI']='Sem_coompanheiro/irmao'
dc.loc[dc.Companheiro_ou_Irmão_a_bordo==1, 'CI']='1_companheiro/irmao'
dc.loc[dc.Companheiro_ou_Irmão_a_bordo>1, 'CI']='>1_compaheiro/irmao'

dc.loc[dc.Pais_ou_filhos_a_bordo==0, 'PF']='Sem_pais/filhos'
dc.loc[dc.Pais_ou_filhos_a_bordo==1, 'PF']='1_pais/filhos'
dc.loc[dc.Pais_ou_filhos_a_bordo>1, 'PF']='>1_pais/filhos'

dc.head()
```

```python
d_c = pd.get_dummies(dc['CI'])
d3 = pd.concat([dc, d_c], axis=1)
d3.head()

d_p = pd.get_dummies(dc['PF'])
d4 = pd.concat([d3, d_p], axis=1)
d4.head()
```
