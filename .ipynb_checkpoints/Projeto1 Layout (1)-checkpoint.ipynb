## Introdução

   O projeto a ser desenvolvido tem como objetivo primário analisar o cenário nacional quanto a distribuição da situação de insegurança alimentar nas mais diversas regiões. Para realizar essa análise, pretende-se primeiramente separar cada região do país e calcular a relação entre o número de habitantes da região em situação de insegurança alimentar e a população total.
   Então, pretende-se destrinchar o perfil da população analisada em relação a sua area de habitação (rural e urbana) e em relação a sua renda, para que mais dados sejam levantados em relação a situação de insegurança alimentar no Brasil.
   Por último, será realizada uma comparação em relação aos dados levantados na PNAD de 2013 e de 2009, para que se compreenda como a situação está evoluindo no Brasil.



## Variáveis
   Para analisar se de fato há uma relação entre esses indicadores, foram escolhidas/criadas as seguintes variáveis:
   
•	Região: Os estados brasileiros foram agrupados em regiões, para que a análise do cenário de insegurança seja feita em relação as regiões, e não em relação aos estados.

•	Pessoas que assinalaram que passam por qualquer tipo de insegurança alimentar: Usada para levantar o pessoal nacional de pessoas que passam por insegurança alimentar no brasil.

•	Pessoas que assinalaram que não passam por qualquer tipo de insegurança alimentar: Usada para levantar o pessoal nacional de pessoas que passam por insegurança alimentar no brasil.

•	Domicílios em zona rural: Agrupamento de todas as áreas rurais da PNAD, utilizada para entender a dinâmica da insegurança alimentar nas zonas do brasil

•	Domicílios em zona urbana: Agrupamento de todas as áreas urbanas da PNAD, mesmo objetivo da variável zona rural.

•	Renda domiciliar per capita: Razão entre a renda domiciliar e o número de habitantes do domicílio.


## Iniciando o projeto: Filtrando variáveis e linhas para serem trabalhadas(2009)

```python
import pandas as pd
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
dados09 = pd.read_csv("DOM2009.csv",sep=",")

#Criando novas variáveis

#Insegurança alimentar(2009)
dados09.loc[(dados09.V2103==1) | (dados09.V2105==1) | (dados09.V2107==1) | (dados09.V2109==1),'Insegurança_Alimentar'] = 'Sim'
dados09.loc[(dados09.V2103==3) & (dados09.V2105==3) & (dados09.V2107==3) & (dados09.V2109==3),'Insegurança_Alimentar'] = 'Não'

#Região
dados09.loc[(dados09.UF >= 10) & (dados09.UF < 20), "Região"] = "Norte"
dados09.loc[(dados09.UF >= 20) & (dados09.UF < 30), "Região"] = "Nordeste"
dados09.loc[(dados09.UF >= 30) & (dados09.UF < 40), "Região"] = "Sudeste"
dados09.loc[(dados09.UF >= 40) & (dados09.UF < 50), "Região"] = "Sul"
dados09.loc[(dados09.UF >= 50) & (dados09.UF < 60), "Região"] = "Centro-Oeste"

#Area do domicílio
dados09.loc[(dados09.V4105 <= 3), "Área_do_domicílio"] = "Urbana"
dados09.loc[(dados09.V4105 >= 4), "Área_do_domicílio"] = "Rural"


#Renda domiciliar percapta
dados09.loc[(dados09.V4621 != 999999999999), "Renda"] = dados09.V4621



#Selecionando algumas colunas iniciais
dados09.UF = dados09.UF.astype("category")
dados09.UF.cat.categories=(['Rondônia', 'Acre', 'Amazonas', 'Roraima', 'Pará', 'Amapá', 'Tocantins', 'Maranhão', 'Piauí', 'Ceará', 'Rio Grande do Norte', 'Paraíba', 'Pernambuco', 'Alagoas', 'Sergipe', 'Bahia', 'Minas Gerais', 'Espírito Santo', 'Rio de Janeiro', 'São Paulo', 'Paraná', 'Santa Catarina', 'Rio Grande do Sul', 'Mato Grosso do Sul', 'Mato Grosso', 'Goiás', 'Distrito Federal'])


#Criando um DataFrame com essas variáveis em 2013
colunas1 = dados09.loc[:,["UF", "Região", "Área_do_domicílio", "Insegurança_Alimentar","Renda"]].dropna(subset=("UF","Região","Área_do_domicílio","Insegurança_Alimentar","Renda"))
colunas1.columns = ["Estado", "Região", "Área_do_domicílio", "Insegurança_Alimentar","Renda"]


#Média da renda domiciliar per-capta das regiões
renda_norte = dados09[(dados09.Região == "Norte")].Renda.mean()
renda_nordeste = dados09[(dados09.Região == "Nordeste")].Renda.mean()
renda_centro_oeste= dados09[(dados09.Região == "Centro-Oeste")].Renda.mean()
renda_sudeste = dados09[(dados09.Região == "Sudeste")].Renda.mean()
renda_sul = dados09[(dados09.Região == "Sul")].Renda.mean()


renda_nordeste
```

## Gerando gráficos para a análise(2009)

```python
#análise da localização geográfica por região do Brasil
fig=plt.figure(figsize=(15,5))

plt.subplot(131)
gregiaoN_IA = colunas1[colunas1.Região == "Norte"].Insegurança_Alimentar.value_counts(True).plot(kind="pie", autopct="%0.1f",legend=True)
plt.title("Região Norte")

plt.subplot(132)
gregiaoNE_IA = colunas1[colunas1.Região == "Nordeste"].Insegurança_Alimentar.value_counts(True).plot(kind="pie", autopct="%0.1f",legend=True)
plt.title("Região Nordeste")

plt.subplot(133)
gregiaoCO_IA = colunas1[colunas1.Região == "Centro-Oeste"].Insegurança_Alimentar.value_counts(True).plot(kind="pie", autopct="%0.1f",legend=True)
plt.title("Região Centro-Oeste")

plt.show()
```

```python
fig=plt.figure(figsize=(15,5))

plt.subplot(131)
gregiaoSE_IA = colunas1[colunas1.Região == "Sudeste"].Insegurança_Alimentar.value_counts(True).plot(kind="pie", autopct="%0.1f",legend=False)
plt.title("Região Sudeste")

plt.subplot(132)
gregiaoS_IA = colunas1[colunas1.Região == "Sul"].Insegurança_Alimentar.value_counts(True).plot(kind="pie", autopct="%0.1f",legend=False)
plt.title("Região Sul")


plt.show()
```

Então foram elaborados gráficos que mostrassem em que área da região (rural ou urbana) a situação de insegurança alimentar é mais frequente

```python
#análise da area da insegurança alimentar
fig=plt.figure(figsize=(15,5))

#localização da insegurança alimentar na região Norte
plt.subplot(131)
colunas1[(colunas1.Região == "Norte") & (colunas1.Insegurança_Alimentar == "Sim")].Área_do_domicílio.value_counts(True).plot(kind="pie", autopct="%0.1f",legend=True)
plt.title("Norte")

#localização da insegurança alimentar na região Nordeste
plt.subplot(132)
plt.title("Nordeste")
colunas1[(colunas1.Região == "Nordeste") & (colunas1.Insegurança_Alimentar == "Sim")].Área_do_domicílio.value_counts(True).plot(kind="pie", autopct="%0.1f",legend=True)

#localização da insegurança alimentar na região Centro Oeste
plt.subplot(133)
plt.title("Centro Oeste")
colunas1[(colunas1.Região == "Centro-Oeste") & (colunas1.Insegurança_Alimentar == "Sim")].Área_do_domicílio.value_counts(True).plot(kind="pie", autopct="%0.1f",legend=True)
```

```python
fig = plt.figure(figsize=(15,5))

#localização da insegurança alimentar na região Sul
plt.subplot(131)
colunas1[(colunas1.Região == "Sul") & (colunas1.Insegurança_Alimentar == "Sim")].Área_do_domicílio.value_counts(True).plot(kind="pie", autopct="%0.1f",legend=True)
plt.title("Sul")

#localização da insegurança alimentar na região Nordeste
plt.subplot(132)
colunas1[(colunas1.Região == "Sudeste") & (colunas1.Insegurança_Alimentar == "Sim")].Área_do_domicílio.value_counts(True).plot(kind="pie", autopct="%0.1f",legend=True)
plt.title("Sudeste")

#Resumo relativo da situação de Insegurança alimentar nas areas rural e urbana

resumo = pd.crosstab((colunas1.Área_do_domicílio), (colunas1.Insegurança_Alimentar))
resumo.plot(kind ="bar", stacked = True)
```

## Filtrando variáveis e linhas para serem trabalhadas(2013)

Para analisar como a insegurança alimentar está geograficamente localizada no brasil, foram elaborados primeiramente gráficos que mostrem a porcentagem de insegurança alimentar em cada região.

```python
import pandas as pd
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
dados = pd.read_csv('DOM2013.csv',sep=',')

#Criando novas variáveis

#Insegurança alimentar(2013)
dados.loc[(dados.V2103==1) | (dados.V2105==1) | (dados.V2107==1) | (dados.V2109==1),'Insegurança_Alimentar'] = 'Sim'
dados.loc[(dados.V2103==3) & (dados.V2105==3) & (dados.V2107==3) & (dados.V2109==3),'Insegurança_Alimentar'] = 'Não'

#Região
dados.loc[(dados.UF >= 10) & (dados.UF < 20), "Região"] = "Norte"
dados.loc[(dados.UF >= 20) & (dados.UF < 30), "Região"] = "Nordeste"
dados.loc[(dados.UF >= 30) & (dados.UF < 40), "Região"] = "Sudeste"
dados.loc[(dados.UF >= 40) & (dados.UF < 50), "Região"] = "Sul"
dados.loc[(dados.UF >= 50) & (dados.UF < 60), "Região"] = "Centro-Oeste"

#Area do domicílio
dados.loc[(dados.V4105 <= 3), "Área_do_domicílio"] = "Urbana"
dados.loc[(dados.V4105 >= 4), "Área_do_domicílio"] = "Rural"


#Renda
dados.loc[(dados.V4621 != 999999999999), "Renda"] = dados.V4621


#Selecionando algumas colunas iniciais
dados.UF = dados.UF.astype("category")
dados.UF.cat.categories=(['Rondônia', 'Acre', 'Amazonas', 'Roraima', 'Pará', 'Amapá', 'Tocantins', 'Maranhão', 'Piauí', 'Ceará', 'Rio Grande do Norte', 'Paraíba', 'Pernambuco', 'Alagoas', 'Sergipe', 'Bahia', 'Minas Gerais', 'Espírito Santo', 'Rio de Janeiro', 'São Paulo', 'Paraná', 'Santa Catarina', 'Rio Grande do Sul', 'Mato Grosso do Sul', 'Mato Grosso', 'Goiás', 'Distrito Federal'])

#Média da renda domiciliar per-capta das regiões
renda_norte = dados[(dados.Região == "Norte")].Renda.mean()
renda_nordeste = dados[(dados.Região == "Nordeste")].Renda.mean()
renda_centro_oeste= dados[(dados.Região == "Centro-Oeste")].Renda.mean()
renda_sudeste = dados[(dados.Região == "Sudeste")].Renda.mean()
renda_sul = dados[(dados.Região == "Sul")].Renda.mean()

#Criando um DataFrame com essas variáveis em 2013
colunas = dados.loc[:,["UF", "Região", "Área_do_domicílio", "Insegurança_Alimentar","Renda"]].dropna(subset=("UF","Região","Área_do_domicílio","Insegurança_Alimentar","Renda"))
colunas.columns = ["Estado", "Região", "Área_do_domicílio", "Insegurança_Alimentar","Renda"]
colunas.tail(8)
```

## Gerando gráficos para a análise(2013)

Para analisar como a insegurança alimentar está geograficamente localizada no brasil, foram elaborados primeiramente gráficos que mostrem a porcentagem de insegurança alimentar em cada região.

```python
#análise da localização geográfica por região do Brasil
fig=plt.figure(figsize=(15,5))

plt.subplot(131)
gregiaoN_IA = colunas[colunas.Região == "Norte"].Insegurança_Alimentar.value_counts(True).plot(kind="pie", autopct="%0.1f",legend=True)
plt.title("Região Norte")

plt.subplot(132)
gregiaoNE_IA = colunas[colunas.Região == "Nordeste"].Insegurança_Alimentar.value_counts(True).plot(kind="pie", autopct="%0.1f",legend=True)
plt.title("Região Nordeste")

plt.subplot(133)
gregiaoCO_IA = colunas[colunas.Região == "Centro-Oeste"].Insegurança_Alimentar.value_counts(True).plot(kind="pie", autopct="%0.1f",legend=True)
plt.title("Região Centro-Oeste")

plt.show()
```

```python
fig=plt.figure(figsize=(15,5))

plt.subplot(131)
gregiaoSE_IA = colunas[colunas.Região == "Sudeste"].Insegurança_Alimentar.value_counts(True).plot(kind="pie", autopct="%0.1f",legend=False)
plt.title("Região Sudeste")

plt.subplot(132)
gregiaoS_IA = colunas[colunas.Região == "Sul"].Insegurança_Alimentar.value_counts(True).plot(kind="pie", autopct="%0.1f",legend=False)
plt.title("Região Sul")


plt.show()
```

Então foram elaborados gráficos que mostrassem em que área da região (rural ou urbana) a situação de insegurança alimentar é mais frequente

```python
#análise da area da insegurança alimentar
fig=plt.figure(figsize=(15,5))

#localização da insegurança alimentar na região Norte
plt.subplot(131)
colunas[(colunas.Região == "Norte") & (colunas.Insegurança_Alimentar == "Sim")].Área_do_domicílio.value_counts(True).plot(kind="pie", autopct="%0.1f",legend=True)
plt.title("Norte")

#localização da insegurança alimentar na região Nordeste
plt.subplot(132)
plt.title("Nordeste")
colunas[(colunas.Região == "Nordeste") & (colunas.Insegurança_Alimentar == "Sim")].Área_do_domicílio.value_counts(True).plot(kind="pie", autopct="%0.1f",legend=True)

#localização da insegurança alimentar na região Centro Oeste
plt.subplot(133)
plt.title("Centro Oeste")
colunas[(colunas.Região == "Centro-Oeste") & (colunas.Insegurança_Alimentar == "Sim")].Área_do_domicílio.value_counts(True).plot(kind="pie", autopct="%0.1f",legend=True)
```
